```python
from kmodes.kprototypes import KPrototypes
```

```python
mark_arrays_AlcoholWrestling = dfbsAlWr1.values
mark_arrays_AlcoholConcert = dfbsAlCo1.values
mark_arrays_RetailWrestling = dfbsReWr1.values
mark_arrays_RetailConcert = dfbsReCo1.values
```

```python
# for alchol wrestling
mark_arrays_AlcoholWrestling[:,1] = mark_arrays_AlcoholWrestling[:,1].astype(float)
mark_arrays_AlcoholWrestling[:,2] = mark_arrays_AlcoholWrestling[:,2].astype(float)
mark_arrays_AlcoholWrestling[:,3] = mark_arrays_AlcoholWrestling[:,3].astype(float)
mark_arrays_AlcoholWrestling[:,9] = mark_arrays_AlcoholWrestling[:,9].astype(float)
```

```python
# for alchol Concert
mark_arrays_AlcoholConcert[:,1] = mark_arrays_AlcoholConcert[:,1].astype(float)
mark_arrays_AlcoholConcert[:,2] = mark_arrays_AlcoholConcert[:,2].astype(float)
mark_arrays_AlcoholConcert[:,3] = mark_arrays_AlcoholConcert[:,3].astype(float)
mark_arrays_AlcoholConcert[:,9] = mark_arrays_AlcoholConcert[:,9].astype(float)
```

```python
# for Retail wrestling
mark_arrays_RetailWrestling[:,1] = mark_arrays_RetailWrestling[:,1].astype(float)
mark_arrays_RetailWrestling[:,2] = mark_arrays_RetailWrestling[:,2].astype(float)
mark_arrays_RetailWrestling[:,3] = mark_arrays_RetailWrestling[:,3].astype(float)
mark_arrays_RetailWrestling[:,9] = mark_arrays_RetailWrestling[:,9].astype(float)
```

```python
# for Retail Concert
mark_arrays_RetailConcert[:,1] = mark_arrays_RetailConcert[:,1].astype(float)
mark_arrays_RetailConcert[:,2] = mark_arrays_RetailConcert[:,2].astype(float)
mark_arrays_RetailConcert[:,3] = mark_arrays_RetailConcert[:,3].astype(float)
mark_arrays_RetailConcert[:,9] = mark_arrays_RetailConcert[:,9].astype(float)
```

```python
mark_arrays_AlcoholWrestling
```

```python
kproto = KPrototypes(n_clusters=3 , verbose = 2 , max_iter = 20) 
clusters = kproto.fit_predict(mark_arrays_AlcoholWrestling,categorical=[0,4,5,6,7,8])
```

```python
print(kproto.cluster_centroids_)
```

```python
cluster_dict = []
for c in clusters:
    cluster_dict.append(c)
```
