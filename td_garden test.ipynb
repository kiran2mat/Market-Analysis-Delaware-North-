```python
from kmodes.kprototypes import KPrototypes
```

```python
dftdAlIc1.head()
```

```python
dftdAlTn1.head()
```

```python
mark_arrays_AlcoholTennis = dftdAlTn1.values
mark_arrays_AlcoholICESH = dftdAlIc1.values
```

```python
mark_arrays_AlcoholTennis[:,1] = mark_arrays_AlcoholTennis[:,1].astype(float)
mark_arrays_AlcoholTennis[:,2] = mark_arrays_AlcoholTennis[:,2].astype(float)
mark_arrays_AlcoholTennis[:,3] = mark_arrays_AlcoholTennis[:,3].astype(float)
mark_arrays_AlcoholTennis[:,9] = mark_arrays_AlcoholTennis[:,9].astype(float)
```

```python
mark_arrays_AlcoholICESH[:,1] = mark_arrays_AlcoholICESH[:,1].astype(float)
mark_arrays_AlcoholICESH[:,2] = mark_arrays_AlcoholICESH[:,2].astype(float)
mark_arrays_AlcoholICESH[:,3] = mark_arrays_AlcoholICESH[:,3].astype(float)
mark_arrays_AlcoholICESH[:,9] = mark_arrays_AlcoholICESH[:,9].astype(float)
```

```python
mark_arrays_AlcoholTennis
```

```python
kproto = KPrototypes(n_clusters=3 , verbose = 2 , max_iter = 20) 
clusters = kproto.fit_predict(mark_arrays_AlcoholTennis,categorical=[0,4,5,6,7,8])
```

```python
print(kproto.cluster_centroids_)
```

```python
cluster_dict = []
for c in clusters:
    cluster_dict.append(c)
```

```python
dftdAlTn['segment'] = cluster_dict
```
