```python
dfhpAlNf1['Product - Cat 3'] = ['Small' if ( x == '12 Ounces' ) else 'Regular' if ( x == '16 Ounces'  or x == '20 Ounces' or x == '19.2 Ounces' or x == 'Regular' ) else 'Large' for x in dfhpAlNf1['Product - Cat 3']]
dfhpAlCo1['Product - Cat 3'] = ['Small' if ( x == '12 Ounces' ) else 'Regular' if ( x == '20 Ounces' or x == '19.2 Ounces'or x == 'Regular' ) else 'Large' for x in dfhpAlCo1['Product - Cat 3']]
```

```python
mark_arrays_AlcoholNFLRG = dfhpAlNf1.values
mark_arrays_AlcoholConcert = dfhpAlCo1.values
mark_arrays_RetailNFLRG = dfhpReNf1.values
mark_arrays_RetailConcert = dfhpReNf1.values
```

```python
# For Alcohol NFLRG

mark_arrays_AlcoholNFLRG[:,1] = mark_arrays_AlcoholNFLRG[:,1].astype(float)
mark_arrays_AlcoholNFLRG[:,2] = mark_arrays_AlcoholNFLRG[:,2].astype(float)
mark_arrays_AlcoholNFLRG[:,3] = mark_arrays_AlcoholNFLRG[:,3].astype(float)
mark_arrays_AlcoholNFLRG[:,9] = mark_arrays_AlcoholNFLRG[:,9].astype(float)
```

```python
# for alchol Concert
mark_arrays_AlcoholConcert[:,1] = mark_arrays_AlcoholConcert[:,1].astype(float)
mark_arrays_AlcoholConcert[:,2] = mark_arrays_AlcoholConcert[:,2].astype(float)
mark_arrays_AlcoholConcert[:,3] = mark_arrays_AlcoholConcert[:,3].astype(float)
mark_arrays_AlcoholConcert[:,9] = mark_arrays_AlcoholConcert[:,9].astype(float)
```

```python
# for Retail NFLRG
mark_arrays_RetailNFLRG[:,1] = mark_arrays_RetailNFLRG[:,1].astype(float)
mark_arrays_RetailNFLRG[:,2] = mark_arrays_RetailNFLRG[:,2].astype(float)
mark_arrays_RetailNFLRG[:,3] = mark_arrays_RetailNFLRG[:,3].astype(float)
mark_arrays_RetailNFLRG[:,9] = mark_arrays_RetailNFLRG[:,9].astype(float)
```

```python
# for Retail Concert
mark_arrays_RetailConcert[:,1] = mark_arrays_RetailConcert[:,1].astype(float)
mark_arrays_RetailConcert[:,2] = mark_arrays_RetailConcert[:,2].astype(float)
mark_arrays_RetailConcert[:,3] = mark_arrays_RetailConcert[:,3].astype(float)
mark_arrays_RetailConcert[:,9] = mark_arrays_RetailConcert[:,9].astype(float)
```

```python
mark_arrays_AlcoholNFLRG
```

```python
kproto = KPrototypes(n_clusters=3 , verbose = 2 , max_iter = 20) 
clusters = kproto.fit_predict(mark_arrays_AlcoholNFLRG,categorical=[0,4,5,6,7,8])
```

```python
print(kproto.cluster_centroids_)
```

```python
cluster_dict = []
for c in clusters:
    cluster_dict.append(c)
```

```python
dfhpAlNf['segment'] = cluster_dict
```
